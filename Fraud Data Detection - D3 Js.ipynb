```python
dates = df1[["IncidentDiscoveredDate", "IncidentID", "Address"]].groupby(["Address","IncidentDiscoveredDate"]).count().reset_index()
dates = dates.rename(columns={"IncidentID":"Number_of_Incidents"})
dates["Number_of_Incidents"] = dates["Number_of_Incidents"].fillna(0)
dates.sort_values(by="Number_of_Incidents")
```

```python
dat = dates["IncidentDiscoveredDate"].astype(str).unique()
add = dates["Address"].astype(str).unique()
columns = [0]*len(dat)
rows = [columns]*len(add)
dff = pd.DataFrame(data=rows, index=add, columns=dat)
dff
```

```python
# Import library
from d3graph import d3graph, vec2adjmat

# Source node names
source = df1.index.tolist()
target = df1["IncidentDiscoveredDate"].astype(str).values.tolist()
weight = df1["Number_of_Incidents"].values.tolist()

# Convert the vector into a adjacency matrix
adjmat = vec2adjmat(source, target, weight, symmetric=False)

# Initialize
d3 = d3graph()
d3.graph(dff.head(117))

r = adjmat.sum(axis=1).to_frame().reset_index()
r = r.sort_values(by=0, ascending=False)
r.columns = ["source", "degree"]

label = r['source'].values

# Set node properties
d3.set_node_properties(label=label, color=label, size=r['degree'].values)

d3.show(figsize=(3000, 3000))
```

```python
def plotD3Graph(file_path, sep, source, destination, weight):
    """
    This method is used to plot a complex Network using D3 Python extension 'd3graph'. 
    If you dont have this library, install using 'pip install d3graph'
    
    :Example:
 
    >>> plotD3Graph('MergedData.csv', '|','Address', 'IncidentDiscoveredDate', 'IncidentID')        
        
    :param arg1: file_path (A complete filepath)
    :type arg1: string
    :param arg2: sep (How csv fields are separated)
    :type arg2: : string 
    :param arg3: source (The nodes from where link will start)
    :type arg3: : string 
    :param arg4: destination (The nodes from where link will end)
    :type arg4: : string 
    :param arg5: weight (The weight/ value of edge between source and destination)
    :type arg5: : string 
    :return: Redirect you to HTML Pages (Also you will have temporary path to HTML file)
    :rtype: HTML File with D3 graph
    :raises: TypeError
    """
    
    # Import library
    from d3graph import d3graph, vec2adjmat
    import pandas as pd
    
    # Read merge data file - which contains features
    df = pd.read_csv(file_path, sep=sep)

    # Fill missing values here
    df[source] = df[source].fillna("Unknown")
    df[destination] = pd.to_datetime(df[destination]).dt.date
    df[weight] = df[weight].fillna("Unknown")
    
    # calculating weight values for network
    dates = df1[[destination, weight]].groupby(destination).count().reset_index()
    dates[weight] = dates[weight].fillna(0)
    
    df1 = df1.merge(dates, on=destination, how="left")
    
    # Source node names
    source = df1[source].astype(str).values.tolist()
    target = df1[destination].astype(str).values.tolist()
    weight = df1[weight].values.tolist()
    
    # Convert the vector into a adjacency matrix
    adjmat = vec2adjmat(source, target, weight)
    
    # Initialize
    d3 = d3graph()
    d3.graph(adjmat)
    d3.show(figsize=(3000, 3000))
```
